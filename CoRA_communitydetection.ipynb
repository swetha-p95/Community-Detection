```python
import networkx as nx
from networkx.algorithms.community.quality import modularity
from collections import defaultdict
import matplotlib.pyplot as plt

# Load the graph
g = nx.read_gml("cora.gml")
g = nx.convert_node_labels_to_integers(g)

```

```python
# Extract communities based on the 'gt' label
communities = defaultdict(list)
for node, data in g.nodes(data=True):
    community_label = data.get("gt")  # Use 'gt' as the partition label
    communities[community_label].append(node)

# Convert communities to a list of sets
community_list = [set(members) for members in communities.values()]

# Calculate modularity
modularity_value = modularity(g, community_list)

print(f"Modularity of the partition: {modularity_value}")# Visualize the communities
gtdict={}
n=0

for i in community_list:
    for j in i:
        gtdict[j] = n
    n+=1
pos = nx.spring_layout(g)  # Positions for all nodes
node_color = [0 for _ in g]
print(len(node_color))
print(gtdict.values())

for i, community in enumerate(community_list):
    for node in community:
        node_color[node] = i
nx.draw(g, pos, node_color=node_color, cmap=plt.cm.rainbow)
plt.title("Cora Ground Truth")
plt.show()
```

```python
import matplotlib.pyplot as plt
import community as community_louvain  # python-louvain library
import numpy as np
# Perform Louvain community detection
louvainpartition = community_louvain.best_partition(g)
print(louvainpartition)
# Print the detected communities
print("Node communities:")
c=[]
for node, community_id in louvainpartition.items():
    # print(community_id,end=" ")
    c.append(community_id)
c=np.array(c)
print(c.max())
c.sort()
print(c)

# Calculate modularity
modularity_value = community_louvain.modularity(louvainpartition, g)

# Display results
print("Modularity of the partition:", modularity_value)

# Visualization of the graph with nodes colored by their community
pos = nx.spring_layout(g)  # Positions for all nodes

# Assign colors to nodes based on their community
colors = [louvainpartition[node] for node in g.nodes()]

plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(g, pos, node_color=colors, cmap=plt.cm.rainbow, node_size=500) #tab10
nx.draw_networkx_edges(g, pos, alpha=0.5)
# nx.draw_networkx_labels(g, pos, font_size=10, font_color="white")

plt.title("Cora Louvain Community Detection")
plt.show()
```

```python
import networkx as nx
import igraph as ig
import leidenalg

# Create a sample NetworkX graph

# Convert NetworkX graph to igraph
G_ig = ig.Graph.TupleList(g.edges(), directed=False)

# Perform Leiden community detection
leidenpartition = leidenalg.find_partition(G_ig, leidenalg.ModularityVertexPartition)
modularity = leidenpartition.quality()
print("Modularity of the partition:", modularity)
# Display the communities
print("Communities:")
leidendict={}

for i, community in enumerate(leidenpartition):
    for j in community:
      leidendict[j]=i

# print(leidenlabels)
# print(partition)
# Map igraph community results back to NetworkX nodes
node_to_community = {G_ig.vs[node]["name"]: i for i, community in enumerate(leidenpartition) for node in community}

# Visualize the communities
import matplotlib.pyplot as plt

pos = nx.spring_layout(g)  # Positions for all nodes
colors = [node_to_community[node] for node in g.nodes()]  # Assign colors based on communities

plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(g, pos, node_color=colors, cmap=plt.cm.rainbow, node_size=500)
nx.draw_networkx_edges(g, pos, alpha=0.5)
# nx.draw_networkx_labels(g, pos, font_size=10, font_color="white")

plt.title("Cora Leiden Community Detection")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.cluster import KMeans
import networkx as nx

# Define the GNN model
class GNN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        # self.conv2 = GCNConv(hidden_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)
    
    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

# Create a sample graph
def create_graph():
    # G = g
    # G = nx.convert_node_labels_to_integers(G)
    # g = G
    # nx.draw(G, with_labels = False)
    edge_index = torch.tensor(list(g.edges(data=False)), dtype=torch.long).t().contiguous()  # Keep only directed edges
    print(g.edges(data=False))
    x = torch.eye(g.number_of_nodes(), dtype=torch.float)  # One-hot encoding for features
    print(edge_index.size())

    print(x.size())
   
    return g, Data(x=x, edge_index=edge_index)
   

# Train the GNN
def train_gnn(model, data, epochs=100, lr=0.01):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)  # Node embeddings
        loss = F.cross_entropy(out @ out.T, torch.eye(out.size(0)))  # Contrastive-like loss
        loss.backward()
        optimizer.step()
    return model

# Perform clustering and calculate modularity
def calculate_modularity(G, embeddings):
    kmeans = KMeans(n_clusters=7, random_state=42).fit(embeddings)
    labels = kmeans.labels_
    partitions = {node: labels[i] for i, node in enumerate(G.nodes())}
    modularity = nx.algorithms.community.modularity(G, [set([node for node, group in partitions.items() if group == i]) for i in set(labels)])
    return partitions, modularity

# Main pipeline
if __name__ == "__main__":
    G, data = create_graph()
    print(data.x.size(1),data.edge_index.size())
    model = GNN(input_dim=data.x.size(1), hidden_dim=64, output_dim=16)
    print(data.x.size())
    print(model)
    model = train_gnn(model, data)
    embeddings = model(data.x, data.edge_index).detach().numpy()
    kmeanspartition, modularity = calculate_modularity(G, embeddings)
    print("Partitions:", kmeanspartition)
    print("Modularity:", modularity)
    print(community_louvain.modularity(kmeanspartition,G))
    # print(kmeanspartition.values())
    
    # Visualization of the graph with nodes colored by their community
    pos = nx.spring_layout(G)  # Positions for all nodes

    # Assign colors to nodes based on their community
    colors = [kmeanspartition[node] for node in G.nodes()]

    plt.figure(figsize=(8, 6))
    nx.draw_networkx_nodes(G, pos, node_color=colors, cmap=plt.cm.rainbow, node_size=500) #tab10
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    # nx.draw_networkx_labels(g, pos, font_size=10, font_color="white")

    plt.title("Cora KMeans Community Detection")
    plt.show()
```
